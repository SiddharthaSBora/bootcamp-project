## Logit models with policies sold as dependent variable

## Load libraries

```python
## For data handling
import pandas as pd
import numpy as np
import re

## For plotting
import matplotlib.pyplot as plt
import seaborn as sns
from stargazer.stargazer import Stargazer
import dataframe_image as dfi

## For Validation
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

## For Performance Measures
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

## Regressions
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LogisticRegression

## This sets the plot style
## to have a grid on a white background
sns.set_style("whitegrid")
```

## Data preprocessing

```python
df = pd.read_csv('../0-data/Root_Insurance_data.csv')
```

```python
### Note: Create Dummies and Intersections
pd.get_dummies(df['click'])
df['click_true'] = pd.get_dummies(df['click'])[True]
### Note: Create dummies for Different Vars
df[["Married", "Single"]]                          = pd.get_dummies(df["Marital Status"])[["M", "S"]]
df[["Insured", "Uninsured", "Unknown_Insurance"]]  = pd.get_dummies(df["Currently Insured"])[["Y", "N", "unknown"]]
df[["SingleDriver", "MultDrivers"]]                = pd.get_dummies(df["Number of Drivers"])[[1, 2]]
df[["OneVehicle", "TwoVehicles", "ThreeVehicles"]] = pd.get_dummies(df["Number of Vehicles"])[[1, 2, 3]]

  # Note: Create rank dummies
df[["r1", "r2", "r3", "r4", "r5"]]                 = pd.get_dummies(df["rank"])[[1,2,3,4,5]]

  # Note: Create Intersections
names1 = ["Married", "Single"]
names2 = ["Insured", "Uninsured", "Unknown_Insurance"]
names3 = ["SingleDriver", "MultDrivers"]
names4 = ["OneVehicle", "TwoVehicles", "ThreeVehicles"]
names = [names1, names2, names3, names4]
### Note: Create continuous intersections
for i in names:
    for j in i:
        df[''.join(["Rank_", j])] = df["rank"]*df[j]

### Note: Create Discrete Intersections
for i in ["r1", "r2", "r3", "r4", "r5"]:
    for j in names:
        for k in j:
            df[''.join([i, "_", k])] = df[i]*df[k]

df.drop(columns = "r5_Unknown_Insurance", axis = 1)
df.columns
```

```python
### Note: Set outcome variable for analysis
outcome_v = "policies_sold"
### Note: Focus on policies sold (for now)
X = df.copy().loc[:,df.columns != outcome_v]
y = df.copy().loc[:,df.columns == outcome_v]
### Note: Create Training Data
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                test_size=.25,
                                                shuffle=True,
                                                random_state=614,
                                                stratify=y)
```

## First we consider the simplest model

At first, we build a logit model with simple features: 'Currently Insured','Number of Vehicles','Number of Drivers', 'Marital Status', 'bid', 'rank'. We consider rank as continous and others as categorical.

Given that policies_sold is a rare event, we try to find a cutoff value for this simple model that maximizes recall and precision. This occurs relatively close to the sample mean of 8%, so we select 0.1 as the cutoff value for assigning poi While this simple model does not perform well, we use this cutoff to identify the optimal functional form; we will revisit the optimal cutoff once we have selected the appropriate functional form. 

```python
# statsmodels formula version allows writing the model in terms of equation 
# Model 1: No frills model with all the explanatory variables, no interaction
df_train = X_train.copy()
df_train["policies_sold"] = y_train.copy()
policies_model_1=smf.logit(formula='policies_sold~ C(Q("Marital Status"))+ C(Q("Number of Vehicles"))+C(Q("Number of Drivers"))+C(Q("Currently Insured"))+rank', data=df_train)
policies_model_1=policies_model_1.fit()

## Write code to calculate the accuracy for any cutoff, then choose your cutoff

## store the predicted probabilities
y_prob = policies_model_1.predict(X_train)
cutoffs = np.arange(0,1.01,.01)

### Note: Illustrate accuracies
accs = []
prec = []
rec  = []
for cutoff in cutoffs:
    y_train_pred = 1*(y_prob > cutoff)
    accs.append(np.sum(y_train_pred == y_train.squeeze())/len(y_train))
    prec.append(precision_score(y_train.squeeze(), y_train_pred))
    rec.append(recall_score(y_train.squeeze(), y_train_pred))
    
plt.figure(figsize=(12,8))
plt.scatter(cutoffs,accs)
plt.xlabel("Cutoff",fontsize=16)
plt.ylabel("Accuracy",fontsize=16)
plt.show()

    
plt.figure(figsize=(12,8))
plt.scatter(cutoffs,prec)
plt.xlabel("Cutoff",fontsize=16)
plt.ylabel("Precision",fontsize=16)
plt.show()

plt.figure(figsize=(12,8))
plt.scatter(cutoffs,rec)
plt.xlabel("Cutoff",fontsize=16)
plt.ylabel("Recall",fontsize=16)
plt.show()
```

## Out of sample performance

```python
# performing predictions on the test datdaset
cutoff=0.1
y_prob = policies_model_1.predict(X_test)
y_pred= 1*(y_prob > cutoff)

from sklearn.metrics import (confusion_matrix, 
                           accuracy_score)  
# confusion matrix
cm = confusion_matrix(y_test, y_pred) 
print ("Confusion Matrix : \n", cm) 
  
# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, y_pred))
```

We got an out of sample accuracy of 71.76% which is not very good. We are worried whether we should consider rank as categorical variable. We examine different functional forms to improve performance and will revisit cutoffs. 

## Adding interaction betweens features
We will try out several models where we add interaction terms to the model, and see if performance improves
 + Model 0: Naive model with discrete features and continuous rank
 + Model 1: Interacting continuous rank with all other features
 + Model 2:
 + Model 2: Discrete rank interacted w/ marital status
 + Model 3: Discrete rank interacted w/ number of vehicles
 + Model 4: Discrete rank interacted w/ drivers
 + Model 5: Discrete rank interacted w/ insured
 + Model 6: Discrete rank interacted w/ mar status/vehicles
 + Model 7: Discrete rank interacted w/ mar status/drivers
 ... ect ect.
 + Model X: Discrete rank interacted w/ all other features

With a cutoff of .10, Model 9 performs best for recall, so we use this model moving forward. This includes discrete rank and intersections between rank and insurance status/multiple drivers.
 

```python
cutoff = 0.1
### Note: Creating lists of variables to make specifications cleaner
ex_vars = ["Married", "TwoVehicles", "ThreeVehicles", "MultDrivers", "Insured", "Unknown_Insurance"]

### Note: Discrete rank
r_disc = []
for i in df.columns:
    if re.search(r"r\d$", i):
        r_disc.append(i)

### Note: Continuous Rank interactions
r_cont = []
for i in df.columns:
    for j in ex_vars:
        if ''.join(["Rank_", j]) in i:
            r_cont.append(i)

### Note: Married interactions
r_mar = []
for i in df.columns:
    if "_Marr" in i:
        if not "Rank_" in i:
            r_mar.append(i)

### Note: Vehicles interactions
r_twoveh = []
for i in df.columns:
    if "_TwoVehicles" in i:
        if not "Rank_" in i:
            r_twoveh.append(i)

r_thrveh = []
for i in df.columns:
    if "_ThreeVehicle" in i:
        if not "Rank_" in i:
            r_twoveh.append(i)
r_veh = r_twoveh + r_thrveh

### Note: Drivers interactions
r_mult = []
for i in df.columns:
    if "_MultD" in i:
        if not "Rank_" in i:
            r_mult.append(i)
            
### Note: Insurance Interactions
r_ins = []
for i in df.columns:
    if "_Insured" in i:
        if not "Rank_" in i:
            r_ins.append(i)
            
r_unk = []
for i in df.columns:
    if "_Unknown" in i:
        if not "Rank_" in i:
            r_unk.append(i)

r_instat = r_ins + r_unk
```

```python
### Note: create specifications
model0 = ex_vars + ["rank"]
model1 = ex_vars + r_cont + ["rank"]
model2 = ex_vars + r_disc
models = [model0, model1, model2]
  # Note: Use powerset function for all combs of rank interactions
def powerset_no_empty(s):
    power_set = []
    x = len(s)
    for i in range(1 << x):
        power_set.append([s[j] for j in range(x) if (i & (1 << j))])
            
    return power_set[1:]

combs = powerset_no_empty(["r_veh", "r_mult", "r_instat", "r_mar"])

for i in combs:
    temp = ex_vars + r_disc
    for j in i:
        temp = temp + eval(j)
    models.append(temp)

```

```python
### Note: Cutoff = 0.1
cutoff = 0.1
### Note: Create Stratified K-Fold
kfold = StratifiedKFold(5,shuffle=True,random_state=440)

X_train_copy = X_train.copy()
y_train_copy = y_train.copy()

n = len(models)
accs = np.zeros((5,n))
precs = np.zeros((5,n))
recs = np.zeros((5,n))
f1s = np.zeros((5,n))

log_reg = LogisticRegression()
### For each fold, fit specification and capture performance metrics
ind_fold = 0
for train_index,test_index in kfold.split(X_train_copy,y_train_copy):
    X_train_train,X_train_test = X_train_copy.iloc[train_index],X_train_copy.iloc[test_index]
    y_train_train,y_train_test = y_train_copy.iloc[train_index],y_train_copy.iloc[test_index]
    
    ind_mod = 0
    for i in models:
        X_tr, X_te = X_train_train[i].to_numpy(), X_train_test[i].to_numpy()
        y_tr, y_te = y_train_train.to_numpy(), y_train_test.to_numpy()
        
        log_reg_clone = clone(log_reg)
        log_reg_clone.fit(X_tr,y_tr.ravel())
        probs = log_reg_clone.predict_proba(X_te)[:,1]
        
        y_pred = 1*(probs > cutoff)
        accs[ind_fold,ind_mod]  = accuracy_score(y_te, y_pred)
        precs[ind_fold,ind_mod] = precision_score(y_te,y_pred)
        recs[ind_fold,ind_mod]  = recall_score(y_te,y_pred)
        f1s[ind_fold,ind_mod]   = f1_score(y_te,y_pred)
        ind_mod += 1
    
    ind_fold += 1

```

```python


# And here if needed
plt.figure(figsize=(10,8))


plt.plot(range(1,n+1),np.mean(precs,axis=0),label="Precision")
plt.plot(range(1,n+1),np.mean(recs,axis=0),label="Recall")
plt.plot(range(1,n+1),np.mean(f1s,axis=0),label="F1 Score")


plt.xlabel("Cutoff", fontsize=16)
plt.ylabel("Mean CV Score", fontsize=16)

plt.legend(fontsize=16)

plt.show()
```

## Analysis of costs

```python
## Note: Create Unique Group Indicators
### Note: Create indicators for unique combinations of observables. We can exploit the fact that we only have 36 unique combinations
  #       of our descriptors (besides rank) to plot and examine heterogeneity in click and buy rates across ranks.
df_working = df.copy()
### Note: Create data frame of unique combinations of our observable variables
temploc  = df_working.columns.get_loc("Married")
temploc2 = df_working.columns.get_loc("ThreeVehicles")
df_temp  = df_working.iloc[:,temploc:temploc2].drop_duplicates().reset_index(drop = True)
df_temp  = df_temp.astype(str)

df_temp.shape

### Note: Create "group" indicators based on combinations of observable characteristics
df_working["Group"] = ""
for i in range(0,df_temp.shape[0]):
    for j in range(0,df.shape[0]):
        tempvec  = ''.join(df_temp.iloc[i,].values)
        tempvec2 = ''.join(df_working.iloc[j,temploc:temploc2].astype(str).values)
        if tempvec == tempvec2:
            df_working.iloc[j,df_working.shape[1]-1] = i
```

```python
### Note: Predict probabilities for sample, reduce to "group means"

  # Note: Predict probs for uncond policies_sold
X_te = df[models[8]].copy().to_numpy()
y_te = df["policies_sold"].copy().to_numpy()
log_reg_clone = clone(log_reg)
log_reg_clone.fit(X_te,y_te.ravel())
probs = log_reg_clone.predict_proba(X_te)[:,1]

df_working["Probs_un"] = probs
  # Note: Group-by Rank and Groups, grab mean probabilities
group_propensities = df_working.groupby(["Group", "rank"])["Probs_un"].mean().unstack()
group_propensities.reset_index(inplace=True)
group_propensities.columns
gp_b = group_propensities.melt(id_vars=["Group"],
                               var_name = "Rank",
                               value_name = "Predicted Buy Rate")

index_names = gp_b[pd.isna(gp_b["Predicted Buy Rate"])].index

  # Note: Transform buy rate into costs
gp_b["CostPerBuy"] = 10/gp_b["Predicted Buy Rate"]
gp = gp_b.drop(index_names)

  # Note: Reshape
gp_reshape = gp.pivot_table(index = "Group",
                      values = "CostPerBuy",
                      columns = "Rank").reset_index()

gp_reshape.columns
gp_reshape["1to2"] = gp_reshape[1] - gp_reshape[2]
gp_reshape["2to3"] = gp_reshape[2] - gp_reshape[3]
gp_reshape["3to4"] = gp_reshape[3] - gp_reshape[4]
gp_reshape["4to5"] = gp_reshape[4] - gp_reshape[5]

gp_reshape["5to4"] = gp_reshape[5] - gp_reshape[4]
gp_reshape["4to3"] = gp_reshape[4] - gp_reshape[3]
gp_reshape["3to2"] = gp_reshape[3] - gp_reshape[2]
gp_reshape["2to1"] = gp_reshape[2] - gp_reshape[1]

### Note: Subtract mean to compare across ranks (note that this weights lower ranks more heavily)
gp_reshape["1to2m"] = gp_reshape["1to2"] - gp_reshape["1to2"].mean()
gp_reshape["2to3m"] = gp_reshape["2to3"] - gp_reshape["2to3"].mean()
gp_reshape["3to4m"] = gp_reshape["3to4"] - gp_reshape["3to4"].mean()
gp_reshape["4to5m"] = gp_reshape["4to5"] - gp_reshape["4to5"].mean()

gp_reshape["5to4m"] = gp_reshape["5to4"] - gp_reshape["5to4"].mean()
gp_reshape["4to3m"] = gp_reshape["4to3"] - gp_reshape["4to3"].mean()
gp_reshape["3to2m"] = gp_reshape["3to2"] - gp_reshape["3to2"].mean()
gp_reshape["2to1m"] = gp_reshape["2to1"] - gp_reshape["2to1"].mean()
gp_reshape["MeanResponsive"]    = -gp_reshape[["1to2m", "2to3m", "3to4m", "4to5m"]].mean(axis=1)
gp_reshape2 = gp_reshape[["Group", "MeanResponsive"]]
df_labels = df_temp.copy()
df_labels.reset_index(inplace = True)
gp_final = gp_reshape2.merge(df_labels,
                            left_on = "Group",
                            right_on = "index")
```

```python
### Display our groups of interest
gp_exp1 = gp_final.sort_values("MeanResponsive", ascending = True).head(10)
dfi.export(gp_exp1, ''.join(["LessResponsive.png"]))

gp_exp2 = gp_final.sort_values("MeanResponsive", ascending = False).head(10)
dfi.export(gp_exp2, ''.join(["MoreResponsive.png"]))
```
