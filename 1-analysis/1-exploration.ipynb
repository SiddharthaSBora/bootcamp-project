### 2) Read in data
Notes
* Bids in the data are flat; there is no variation in the bid-price in the data, but rank varies across observations. Given that this data was generated from a first-price auction, Rank contains information on the relative bid by Root. We know whether other firms bid higher or lower than 10 for each observation. 
* Both "Number of Vehicles" and "Number of Drivers" take only a few unique values; these variables might be best treated as categorical
* There are only 3x3x2x2 = 36 unique combinations of consumer characteristics. Counts within unique groups range from 326 to 256
* Number of Vehicles, Number of Drivers, and Currently Insured are balanced in the sample. However, we observe approximately 50% more consumers in Rank 5 as Rank 1, suggesting we might not see all groups exposed to all ranks. 
* Full sample click probabilities are approximately 18% and sale probabilities are around 8%

### 3) Create Variables of Interest

```python
df_working = df.copy()

### Note: Create categorical (factor) versions of descriptors
varlist = ["Currently Insured", "Number of Vehicles", "Number of Drivers", "Marital Status"]
for i in varlist:
    j             = i + "_Cat"
    df_working[j] = pd.Categorical(df_working[i])
```

```python
### Note: Create dummies for Different Vars
df_working[["Married", "Single"]]                          = pd.get_dummies(df_working["Marital Status"])[["M", "S"]]
df_working[["Insured", "Uninsured", "Unknown_Insurance"]]  = pd.get_dummies(df_working["Currently Insured"])[["Y", "N", "unknown"]]
df_working[["SingleDriver", "MultDrivers"]]                = pd.get_dummies(df_working["Number of Drivers_Cat"])[[1, 2]]
df_working[["OneVehicle", "TwoVehicles", "ThreeVehicles"]] = pd.get_dummies(df_working["Number of Vehicles"])[[1, 2, 3]]

df_working.columns
```

```python
### Note: Create indicators for unique combinations of observables. We can exploit the fact that we only have 36 unique combinations
  #       of our descriptors (besides rank) to plot and examine heterogeneity in click and buy rates across ranks.

### Note: Create data frame of unique combinations of our observable variables
temploc  = df_working.columns.get_loc("Married")
temploc2 = len(df_working.columns)-1
df_temp  = df_working.iloc[:,temploc:temploc2].drop_duplicates().reset_index(drop = True)
df_temp  = df_temp.astype(str)

df_temp.shape

### Note: Create "group" indicators based on combinations of observable characteristics
df_working["Group"] = ""
for i in range(0,df_temp.shape[0]):
    for j in range(0,df.shape[0]):
        tempvec  = ''.join(df_temp.iloc[i,].values)
        tempvec2 = ''.join(df_working.iloc[j,temploc:temploc2].astype(str).values)
        if tempvec == tempvec2:
            df_working.iloc[j,df_working.shape[1]-1] = i
```

```python
### Note: Full sample click and buy probabilities
mean_click = df_working[df_working.click == 1]["click"].count()/df_working.shape[0]
mean_buy   = df_working[df_working.policies_sold == 1]["policies_sold"].count()/df_working.shape[0]
mean_condb = df_working[df_working.policies_sold == 1]["policies_sold"].count()/df_working[df_working.click == 1].shape[0]

### Note: Table of value counts
varlist = ["Currently Insured", "Number of Vehicles", "Number of Drivers", "Marital Status", "rank", "click", "policies_sold"]
df_vc = pd.DataFrame({"Value": [],
                      "Count": [],
                      "Variable": []})
for i in varlist:
    df_vc_t = df_working.value_counts(i).sort_index().to_frame()
    df_vc_t.reset_index(inplace = True)
    df_vc_t.columns = ["Value", "Count"]
    df_vc_t["Variable"] = i
    df_vc = df_vc.append(df_vc_t)
    display(df_vc_t)
```

```python
### Note: Plot across different groups for rank for click and for policies sold
groups = df_working.Group.value_counts()
groups_df = groups.to_frame()
groups_df.reset_index(inplace=True)
groups_df.columns = ["Group", "Count"]

group_propensities = df_working.groupby(["Group", "rank"])["click"].mean().unstack()
group_propensities.reset_index(inplace=True)
group_propensities.columns
gp_c = group_propensities.melt(id_vars=["Group"],
                               var_name = "Rank",
                               value_name = "Success Rate")
gp_c["Outcome"] = "Click"

index_names = gp_c[pd.isna(gp_c["Success Rate"])].index
gp_c.drop(index_names)

### Note: policies sold group propensities
group_propensities = df_working.groupby(["Group", "rank"])["policies_sold"].mean().unstack()
group_propensities.reset_index(inplace=True)
group_propensities.columns
gp_b = group_propensities.melt(id_vars=["Group"],
                               var_name = "Rank",
                               value_name = "Success Rate")

gp_b["Outcome"] = "Policy Sold"
index_names = gp_b[pd.isna(gp_b["Success Rate"])].index
gp_b.drop(index_names)

### Note: Conditional sale rates
df_working1 = df_working.drop(df_working[df_working.click == 0].index)
group_propensities = df_working1.groupby(["Group", "rank"])["policies_sold"].mean().unstack()
group_propensities.reset_index(inplace=True)
group_propensities.columns
gp_cb = group_propensities.melt(id_vars=["Group"],
                               var_name = "Rank",
                               value_name = "Success Rate")

gp_cb["Outcome"] = "Policy Sold | Click"
index_names = gp_cb[pd.isna(gp_cb["Success Rate"])].index
gp_cb.drop(index_names)

### Note: Append the three datasets together
gp = gp_c.append(gp_b, ignore_index = True)
gp = gp.append(gp_cb, ignore_index = True)

gp2 = gp.merge(groups_df,
               how = "left",
               on = "Group")

plot = sns.catplot(x = "Rank",
                   y = "Success Rate",
                   col = "Outcome",
                   hue = "Group",
                   data = gp,
                   kind = "swarm")

plot._legend.remove()

plot.savefig("GroupPropensities.png")
```

```python
# ### Note: Success Rates across different variables
var_l = ["Number of Vehicles", "Currently Insured", "Number of Drivers", "Marital Status"]

for i in var_l:
    fig, ax =plt.subplots(1,3,
                          figsize = (12,8))
    plot2 = sns.lineplot(x = "rank",
                         y = "click",
                         hue = i,
                         data = df_working,
                         ax = ax[0])
    plot2.set_ylim([0,1])
    plot2.set_title("Click")

    plot3 = sns.lineplot(x = "rank",
                        y = "policies_sold",
                        hue = i,
                        data = df_working,
                        ax = ax[1])
    plot3.set_ylim([0,1])
    plot3.set_title("Policy Sold")

    plot4 = sns.lineplot(x = "rank",
                         y = "policies_sold",
                         hue = i,
                         data = df_working[df_working.click == 1],
                         ax = ax[2])
    plot4.set_ylim([0,1])
    plot4.set_title("Policy Sold | Click")
    fig.suptitle(''.join([i, " Engagement"]))
    fig.savefig(''.join([i, "_Propensities.png"]))
    fig.clear()      
```
